```python


categorical_features = [
    "Title (Company)",
    "Description",
    "Jurisdiction",
    "LINDDUN categories",
    "Threat Actor",
    "Techniques Used",
    "Data Protection State",
    "Root Cause (Why)",
    "Asset Type",
    "Source"
]

unknown_labels = [
    "Unknown",
    "No Label",
    "No description available",
    "Unknown Entity",
    "N/A",
    ""
]

november_data_copy = november_data.copy(deep=True)

for col in categorical_features:
    november_data_copy[col] = (
        november_data_copy[col]
        .replace(unknown_labels, "Unknown")
        .fillna("Unknown")
    )
```

```python
numerical_features = [
    "Data exposure score",
    "Identification score",
    "Data Sensitivity score",
    "User Control score"
]

for col in numerical_features:
    november_data_copy[col] = november_data_copy[col].replace(-1, np.nan)
```

```python
cat_results = []

for col in categorical_features:
    group_col = col + "_Group"

    november_data_copy[group_col] = november_data_copy[col].apply(
        lambda x: "Unknown" if x == "Unknown" else "Known"
    )

    counts = november_data_copy[group_col].value_counts()
    percentages = november_data_copy[group_col].value_counts(normalize=True) * 100
    
    cat_results.append({
        "Feature": col,
        "Known_Count": counts.get("Known", 0),
        "Unknown_Count": counts.get("Unknown", 0),
        "Known_%": round(percentages.get("Known", 0), 2),
        "Unknown_%": round(percentages.get("Unknown", 0), 2)
    })
```

```python
num_results = []

for col in numerical_features:
    group_col = col + "_Group"
    
    november_data_copy[group_col] = november_data_copy[col].apply(
        lambda x: "Unknown" if pd.isna(x) else "Known"
    )

    counts = november_data_copy[group_col].value_counts()
    percentages = november_data_copy[group_col].value_counts(normalize=True) * 100
    
    num_results.append({
        "Feature": col,
        "Known_Count": counts.get("Known", 0),
        "Unknown_Count": counts.get("Unknown", 0),
        "Known_%": round(percentages.get("Known", 0), 2),
        "Unknown_%": round(percentages.get("Unknown", 0), 2)
    })
```

```python
known_unknown_categorical = pd.DataFrame(cat_results)
known_unknown_numerical = pd.DataFrame(num_results)

print(known_unknown_categorical)
print(known_unknown_numerical)
```

```python
import pandas as pd
import plotly.express as px
from dash import Dash, dcc, html, Input, Output

# -------------------------
# Load and prepare data
# -------------------------

df = november_data_copy.copy()

def known_unknown(x):
    if pd.isna(x):
        return "Unknown"
    x = str(x).strip().lower()
    if x in ["unknown", "no label", "not available"]:
        return "Unknown"
    return "Known"

df["ThreatActor_KU"] = df["Threat Actor"].apply(known_unknown)
df["Technique_KU"] = df["Techniques Used"].apply(known_unknown)
df["Protection_KU"] = df["Data Protection State"].apply(known_unknown)
df["RootCause_KU"] = df["Root Cause (Why)"].apply(known_unknown)

# -------------------------
# Initialize app
# -------------------------
app = Dash(__name__)

# -------------------------
# App layout
# -------------------------
app.layout = html.Div(
    style={"padding": "20px"},
    children=[

        html.H1("Cybersecurity & Privacy Incident Dashboard"),

        html.Label("Filter Data Visibility"),
        dcc.Dropdown(
            id="known-filter",
            options=[
                {"label": "All Incidents", "value": "all"},
                {"label": "Known Only", "value": "known"},
                {"label": "Unknown Only", "value": "unknown"},
            ],
            value="all",
            clearable=False,
            style={"width": "300px"}
        ),

        html.Br(),

        dcc.Graph(id="threat-tech-graph"),
        dcc.Graph(id="protection-sensitivity-graph"),
        dcc.Graph(id="motivation-asset-graph"),
    ]
)

# -------------------------
# Callback
# -------------------------
@app.callback(
    Output("threat-tech-graph", "figure"),
    Output("protection-sensitivity-graph", "figure"),
    Output("motivation-asset-graph", "figure"),
    Input("known-filter", "value")
)
def update_dashboard(filter_value):

    filtered_df = df.copy()

    if filter_value == "known":
        filtered_df = filtered_df[
            (filtered_df["ThreatActor_KU"] == "Known") &
            (filtered_df["Technique_KU"] == "Known") &
            (filtered_df["Protection_KU"] == "Known") &
            (filtered_df["RootCause_KU"] == "Known")
        ]

    elif filter_value == "unknown":
        filtered_df = filtered_df[
            (filtered_df["ThreatActor_KU"] == "Unknown") |
            (filtered_df["Technique_KU"] == "Unknown") |
            (filtered_df["Protection_KU"] == "Unknown") |
            (filtered_df["RootCause_KU"] == "Unknown")
        ]

    # -------------------------
    # Plot 1: Threat Actor vs Techniques Used
    # -------------------------
    ta_tech = (
        filtered_df
        .groupby(["Threat Actor", "Techniques Used"])
        .size()
        .reset_index(name="Count")
    )

    fig1 = px.bar(
        ta_tech,
        x="Threat Actor",
        y="Count",
        color="Techniques Used",
        title="Threat Actor vs Techniques Used"
    )

    # -------------------------
    # Plot 2: Protection State vs Data Sensitivity
    # -------------------------
    fig2 = px.box(
        filtered_df,
        x="Data Protection State",
        y="Data Sensitivity score",
        title="Data Protection State vs Data Sensitivity Score"
    )

    # -------------------------
    # Plot 3: Motivation vs Asset Type (Normalized)
    # -------------------------
    mot_asset = pd.crosstab(
        filtered_df["Root Cause (Why)"],
        filtered_df["Asset Type"],
        normalize="index"
    ).reset_index()

    mot_asset = mot_asset.melt(
        id_vars="Root Cause (Why)",
        var_name="Asset Type",
        value_name="Percentage"
    )

    mot_asset["Percentage"] *= 100

    fig3 = px.bar(
        mot_asset,
        x="Root Cause (Why)",
        y="Percentage",
        color="Asset Type",
        title="Motivation vs Asset Type Involved"
    )

    return fig1, fig2, fig3


# -------------------------
# Run app
# -------------------------
if __name__ == "__main__":
    app.run(debug=True)
```
